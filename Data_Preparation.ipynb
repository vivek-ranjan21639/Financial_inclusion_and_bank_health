## Finanacial Inclusion

### Access

```python
df["scb_branches_per_lakh"] = (df["scb_offices"] / df["total_population"]) * 100000
```

```python
df["rrb_branches_per_lakh_rural"] = (df["rrb_branches"] / df["rural_population"]) * 100000
```

### Usage

```python
df["scb_deposit_per_capita"] = df["scb_deposits"] / df["total_population"]
df["scb_credit_per_capita"] = df["scb_credit"] / df["total_population"]
df["rrb_credit_per_capita"] = df["rrb_credit"] / df["rural_population"]
```

### Bank Health Indicators

```python
df["cd_gap"] = df["cd_scb_sanction"] - df["cd_scb_utilisation"]

def cd_category(x):
    if x < 60:
        return "Low CD"
    elif x <= 90:
        return "Balanced CD"
    else:
        return "High CD"

df["cd_scb_category"] = df["cd_scb_utilisation"].apply(cd_category)
df["cd_rrb_category"] = df["cd_rrb"].apply(cd_category)
```

### Credit Structure & Risk

```python
df["agri_share"] = df["agri_credit"] / df["scb_credit"]
df["industry_share"] = df["industry_credit"] / df["scb_credit"]
df["personal_loan_share"] = df["personal_loans"] / df["scb_credit"]

df["credit_concentration"] = df[
    ["agri_share", "industry_share", "personal_loan_share"]
].max(axis=1)
```

### Composite Inclusion Index

```python
inclusion_vars = [
    "scb_branches_per_lakh",
    "scb_deposit_per_capita",
    "scb_credit_per_capita"
]
```

```python
inclusion_vars = [
    "scb_branches_per_lakh",
    "scb_deposit_per_capita",
    "scb_credit_per_capita"
]

df[[f"{v}_z" for v in inclusion_vars]] = df[inclusion_vars].apply(zscore)

df["inclusion_index"] = df[
    [f"{v}_z" for v in inclusion_vars]
].mean(axis=1)
```

### State Typology

```python
income_median = df["per_capita_nsdpconstant_prices"].median()
inclusion_median = df["inclusion_index"].median()

def state_type(row):
    if row["per_capita_nsdpconstant_prices"] >= income_median and row["inclusion_index"] >= inclusion_median:
        return "High Income  High Inclusion"
    elif row["per_capita_nsdpconstant_prices"] < income_median and row["inclusion_index"] >= inclusion_median:
        return "Low Income  High Inclusion"
    elif row["per_capita_nsdpconstant_prices"] >= income_median and row["inclusion_index"] < inclusion_median:
        return "High Income  Low Inclusion"
    else:
        return "Low Income  Low Inclusion"

df["state_type"] = df.apply(state_type, axis=1)
```
